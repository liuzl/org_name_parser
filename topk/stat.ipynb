```python
import pandas as pd
```

```python
data = pd.read_csv("output_suffix.txt",sep=' ',header=None, names=["word","cnt","err"])
```

```python
filtered = data[data["cnt"]-data["err"]>=10]
#filtered = filtered[filtered['cnt']>100]
```

```python
m = {}
def f1():
    for index, row in filtered.iterrows():
        row['sum'] = row['cnt']
        m[row['word']] = row
def f2():
    for index, row in filtered.iterrows():
        for i in range(1, len(row['word'])):
            w = row['word'][i:]
            if w in m:
                m[w]['sum'] -= row['cnt']
                if m[w]['sum']/m[w]['cnt']<0.3:
                    del m[w]
f1()
f2()
```

```python
for k, v in m.items():
    print(k, v['cnt'])
```
