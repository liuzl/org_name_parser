```python
import pandas as pd
```

```python
data = pd.read_csv("output_suffix.txt",sep=' ',header=None, names=["word","cnt","err"])
```

```python
filtered = data[data["cnt"]-data["err"]>=10]
```

```python
def f1(d, m):
    for index, row in d.iterrows():
        row['sum'] = row['cnt']
        m[row['word']] = row
def f2(d, m):
    for index, row in d.iterrows():
        for i in range(1, len(row['word'])):
            w = row['word'][i:]
            if w in m:
                m[w]['sum'] -= row['cnt']
                if m[w]['sum']/m[w]['cnt']<0.3:
                    del m[w]
```

```python
suffixes = {}
f1(filtered, suffixes)
f2(filtered, suffixes)
```

```python
def output(f, m):
    with open(f, "w") as out:
        for k, v in m.items():
            out.write("%s\t%d\n" % (v['word'], v['cnt']-v['err']))
```

```python
output("suffixes.txt", suffixes)
```
